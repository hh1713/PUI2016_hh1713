```python
def create_chisq_size(df, n):
    return np.random.chisquare(df, size = n).mean()
```

```python
np.random.seed(229)
df = 100

chisq = np.random.chisquare(df, size = 100)
chisq_mean = [create_chisq_size(df, n) for n in np.linspace(10, 1000, 100)] #100 numbers that go from 10 to 1000
print(chisq_mean)
```
